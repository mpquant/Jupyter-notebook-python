# Matplotlib 

- 它能让使用者很轻松地将数据图形化，并且提供多样化的输出样式

- 与pandas、numpy搭配使用，让结果以更真实实的方式显示


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 一个随机的线性显示图显示
df = pd.DataFrame(np.random.randn(10,2),index=pd.date_range('2022/01/01',
   periods=10), columns=list('AB'))
df.plot()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 简单的抛物线
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
```
